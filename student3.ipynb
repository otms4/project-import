```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df = pd.read_csv(r"C:\Users\EL10_gazy\Downloads\kaggle data science\IRIS.csv")
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Create a figure with a subplot for each numerical feature
plt.figure(figsize=(8, 6))

# Loop through the numerical columns and plot
for i, col in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=15, color='skyblue')
    plt.title(f'Distribution of {col.replace("_", " ").title()}', fontsize=12)
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
